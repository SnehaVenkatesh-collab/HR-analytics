### GENDER

```python
ax = sns.countplot(x = 'Gender',data = df)
sns.set(rc={'figure.figsize':(8,3)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
income_gen = df.groupby(['Gender'], as_index=False)['MonthlyIncome'].sum().sort_values(by='MonthlyIncome', ascending=False)

sns.barplot(x = 'Gender',y= 'MonthlyIncome' ,data = income_gen )
```

### Job Satisfaction

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x = 'JobSatisfaction',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

### Work Life Balance

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x = 'WorkLifeBalance',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

### Business Travel
    

```python
ax = sns.countplot(data = df , x = 'BusinessTravel')
for bars in ax.containers:
  ax.bar_label(bars)
```

### JOB ROLES

```python

palette = sns.color_palette("dark")
ax = sns.countplot(x = 'JobRole',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
jobrole_gen = df.groupby(['JobRole'], as_index=False)['MonthlyIncome'].sum().sort_values(by='MonthlyIncome', ascending=False)

sns.barplot(x = 'JobRole',y= 'MonthlyIncome' ,data = jobrole_gen )
```

### DEPARTMENT

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Department')

for bars in ax.containers:
    ax.bar_label(bars)
```

### MARITAL STATUS

```python
ax = sns.countplot(data = df, x = 'MaritalStatus')

for bars in ax.containers:
    ax.bar_label(bars)
```

### ATTRITION

```python
ax = sns.countplot(data = df , x = 'Attrition')
```

```python
grouped_df = df.groupby(['BusinessTravel', 'Attrition']).size().reset_index(name='Count')
sns.barplot(data=grouped_df, x='BusinessTravel', y='Count', hue='Attrition')
plt.title('Attrition by BusinessTravel')
plt.xlabel('BusinessTravel')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

```python
grouped_df = df.groupby(['Gender', 'Attrition']).size().reset_index(name='Count')
sns.barplot(data=grouped_df, x='Gender', y='Count', hue='Attrition')
plt.title('Attrition by Gender ')
```

```python
grouped_df = df.groupby(['MaritalStatus', 'Attrition']).size().reset_index(name='Count')
sns.barplot(data=grouped_df, x='MaritalStatus', y='Count', hue='Attrition')
plt.title('Attrition by MaritalStatus')
plt.xlabel('MaritalStatus')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

```python
grouped_df = df.groupby(['JobRole', 'Attrition']).size().reset_index(name='Count')
sns.barplot(data=grouped_df, x='JobRole', y='Count', hue='Attrition')
plt.title('Attrition by JobRole')
plt.xlabel('JobRole')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

```python
grouped_df = df.groupby(['JobSatisfaction', 'Attrition']).size().reset_index(name='Count')
sns.barplot(data=grouped_df, x='JobSatisfaction', y='Count', hue='Attrition')
plt.title('Attrition by JobSatisfaction')
plt.xlabel('JobSatisfaction')
plt.ylabel('Count')
plt.legend(title='Attrition')
```

```python
grouped_df = df.groupby(['WorkLifeBalance', 'Attrition']).size().reset_index(name='Count')
sns.barplot(data=grouped_df, x='WorkLifeBalance', y='Count', hue='Attrition')
plt.title('Attrition by WorkLifeBalance')
plt.xlabel('WorkLifeBalance')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```
